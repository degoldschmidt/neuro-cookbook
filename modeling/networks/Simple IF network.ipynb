# Simple integrate-and-fire network

$\tau_m \frac{dU_i}{dt} = âˆ’u_i +RI_i(t)$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# constants
N = 100 # number of neurons
dt = 0.001 # integration time
tau_m = 10 # membrane time const.
R = 1+np.random.randn(N) # input resistance
```

```python
# containers
time = np.arange(0,100,dt)
in_signal = 0.2*np.random.randn(time.shape[0]) + .2*np.sin(0.5*time) + .1*np.cos(0.02*time)
U = np.zeros((N,time.shape[0])) 
```

```python
plt.plot(time, in_signal)
plt.show()
```

```python
for n,t in enumerate(time):
    dU = (-U[:,n-1] + R*in_signal[n])
    U[:,n] = U[:,n-1] + (dt/tau_m)*dU
```

```python
plt.plot(time, U[0])
plt.plot(time, U[1])
plt.show()
```
